```python
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
import pandas as pd

# Assuming 'data.csv' is the name of the uploaded file
file_path = '/content/Rohit_Sharma_Centuries.csv'

# Try reading the file with different encodings
encodings = ['latin1', 'iso-8859-1', 'cp1252']
df = None

for enc in encodings:
    try:
        df = pd.read_csv(file_path, encoding=enc)
        print(f"File successfully read with encoding: {enc}")
        break
    except UnicodeDecodeError:
        print(f"Failed to read file with encoding: {enc}")

if df is not None:
    # Display the dataframe
    print(df.head(3))
else:
    print("Failed to read the file with all attempted encodings.")

```

# **plot()**
- It makes charts - defaults is line plot
- More graphs with kinda='line','bar','scatter','hist',etc
# **Line Plot**
        syntax: plot() or plot.line()
        

```python
df.plot()
```

# **Scatter Plot**
           syntax: plot(kind='scatter',x,y) or plot.scatter(x,y)

```python
df.plot(kind='scatter', x='Score',y='Strike Rate')
```

```python
df.plot.scatter(x='Position',y='Score')
```

# **Histogram**
        syntax: plot(kind='hist') or plot.hist()

```python
df.plot(kind='hist')
```

# **Area Plot**
       syntax: plot(kind='area') or plot.area()

```python
df.plot.area()
```

# **Bar chart**
      syntax: plot(kind='bar) or plot.bar()
              barh() for horizontal bar

```python
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh()
```

# **Box plot**
     syntax: plot(kind='box') or plot.box()

```python
df.plot.box()
```

# **Hexagonal Binning Plot**
     syntax: plot.hexbin()

```python
df.plot.hexbin(x='Position',y='Innings',gridsize=18)
```

# **Histogram Plot**
    syntax: plot.hist()

```python
df.plot.hist()
```

# **Destiny Estimate Plot**
      syntax: plot.kde() or plot.density()
      kernel density estimate charts

```python
df.plot.kde()
```

# **Pie Plot**
   syntax: pie.plot()

```python
df.plot.pie(y='Strike Rate',autopct='%1.1f%%')
```
